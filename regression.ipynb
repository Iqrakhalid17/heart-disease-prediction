```python
y = data['target'] # dependant                       # splitting the data into dependant and independant variables
x = data.drop(['target'], axis = 1) #independant
```

```python
from sklearn.model_selection import train_test_split         # importing library
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.33, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression  # importing linear regression model
```

```python
regressor = LinearRegression()    #fitting the model into x and y train
regressor.fit(x_train, y_train)
```

```python
y_pred_test = regressor.predict(x_test)   #predicting the test and train
y_pred_train = regressor.predict(x_train)   
```

```python
print(f'Coefficient: {regressor.coef_}')          # identifying the co-efficient and intercept to put in linear equation
print(f'Intercept: {regressor.intercept_}')
print('y = ', regressor.intercept_, '+', regressor.coef_, '* X')
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
print('MSE = ', mean_squared_error(y_test, y_pred_test ))       # finding mean squred error, r2 score and root mean square error
print('R2 = ', r2_score(y_test, y_pred_test )) 
print('RMSE = ', np.sqrt(mean_squared_error(y_test, y_pred_test )))
```

```python
from sklearn.neighbors import KNeighborsClassifier    # importing knn classifier model

knn = KNeighborsClassifier(n_neighbors=7)
knn.fit(x_train,y_train)
y_pred_knn=knn.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
score_knn = round(accuracy_score(y_pred_knn,y_test)*100,2)   # accuracy for KNN model

print("The accuracy score achieved using KNN is: "+str(score_knn)+" %")
```
